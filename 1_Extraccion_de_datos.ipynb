```python
#Cargar archivo scv desde equipo
from google.colab import files

RGO=pd.read_csv("CountryTable.csv")
#RGO
```

```python
#Imprimimos las primeras filas o registros del Dataframe

RGO.head(5)
```

```python
#*FILTROS BASICOS PARA VARIABLES CUALITATIVAS*
#Filtro por objeto
filtro1=RGO[RGO["continent"]=="North America"]
filtro1
```

```python
#Filtro por filas
#filtro2= RGO.iloc[300:302,:]
#filtro2=RGO.iloc[[1,2,3],:]#Filas no consecutivas
filtro2=RGO.iloc[1:11,:]#Filas consecutivas
filtro2
```

```python
#Filtro por columnas
filtro3=RGO.iloc[:,4:7] #columnas consecutivas
#filtro3=RGO.iloc[:,[1,4,6]] #columnas no consecutivas
filtro3
```

```python
#filtro por
RGO=pd.read_csv("CountryTable.csv", index_col=2) #columna indice, empezara con la columna continent
RGO
```

```python
#Filtro de filas contra columnas, que me arroje solo las filas de la columna indice
filtro4=RGO.loc[["Asia","Africa"],["population"]]
filtro4
```

```python
#Filtro por cabecera
filtro5=RGO.head(6)#muestra las primeras filas hasta el numero marcado
filtro5
```

Filtros Basicos para variables cuantitativas

```python
#Filtro por comparacion
filtro6=RGO[RGO["life_expectancy"]>60]
filtro6
```

Filtros Basicos con Conectores Propios

```python
#Filtro Y
filtro7=RGO[(RGO["life_expectancy"]>60) & (RGO["region"]=="Caribbean")]
filtro7
```

```python
#Filtro o
filtro8=RGO[(RGO["life_expectancy"]>60) | (RGO["region"]=="Caribbean")] #Te da todos los valores que cumplen con una o la otra
filtro8
```

```python
#Filtro not
filtro9=RGO[~(RGO["life_expectancy"]>=60) & ~(RGO["region"]=="Caribbean")]#excluye todos los valores con dichas caracteristicas
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Filtro1.csv")
filtro2.to_csv("Filtro2.csv")
filtro3.to_csv("Filtro3.csv")
filtro4.to_csv("Filtro4.csv")
filtro5.to_csv("Filtro5.csv")
filtro6.to_csv("Filtro6.csv")
filtro7.to_csv("Filtro7.csv")
filtro8.to_csv("Filtro8.csv")
filtro9.to_csv("Filtro9.csv")
```
